```python
import pandas as pd
```

```python
df = pd.read_csv('./Advertising.csv')
```

```python
df
```

```python
df.describe()
```

```python
X = df.drop('sales', axis = 1)
y = df['sales']
```

```python
# Train | Validation | Hold out test
# 70%   | 15%        | 15%

# 1st split - 70/30
# 2nd split 50/50 (on 30% test set)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)
```

```python
X_validation, X_holdout_test, y_validation, y_holdout_test = train_test_split(X_test, y_test, test_size = 0.5, random_state = 101)
```

```python
len(X)
```

```python
len(X_train)
```

```python
len(X_validation)
```

```python
len(X_holdout_test)
```

# Model Training

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators = 3, random_state = 101)
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
validation_predictions = model.predict(X_validation)
```

```python
validation_predictions
```

```python
mean_absolute_error(y_validation, validation_predictions)
```

```python
mean_squared_error(y_validation, validation_predictions) ** 0.5 #RMSE STD
```

```python
df.describe()
```

```python
df.describe()['sales'] # MAE : 0.85, RMSE : 1.1
```

# Model 2 with increased no. of estimators


```python
model_2 = RandomForestRegressor(n_estimators = 30, random_state = 101)
```

```python
model_2.fit(X_train, y_train)
```

```python
validation_predictions_2 = model_2.predict(X_validation)
```

```python
validation_predictions_2
```

```python
mean_absolute_error(y_validation, validation_predictions_2)
```

```python
mean_squared_error(y_validation, validation_predictions_2) ** 0.5 # RMSE STD
```

```python
df.describe()['sales'] # MAE : 0.65, RMSE : 085
```

# Final Performance Metrics (HOLDOUT SET)

```python
holdout_predictions = model_2.predict(X_holdout_test)
```

```python
mean_absolute_error(y_holdout_test, holdout_predictions) # MAE
```

```python
mean_squared_error(y_holdout_test, holdout_predictions)**0.5 # RMSE
```

```python
final_model = RandomForestRegressor(n_estimators = 30, random_state = 101)
```

```python
final_model.fit(X, y)
```
